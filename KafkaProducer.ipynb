```python
import pandas as pd
from kafka import KafkaConsumer, KafkaProducer
from time import sleep
import json
```

## Producer

```python
# Create teh producer

producer = KafkaProducer(
                          bootstrap_servers = ['**.**.**.**:9092'],
                          value_serializer = lambda x:
                          json.dumps(x).encode('utf-8')
)
```

```python

# Teste if the producer work well
producer.send(
              'demo_01',
              value = "{'name':'hahah'}"
              )
```

```python
# Get the data from HTTP request(from my github)

import requests
from io import StringIO

response = requests.get("https://raw.githubusercontent.com/mooohh/Kafka-Market-Stream/refs/heads/master/Data/indexProcessed.csv")

if response.status_code == 200:
    # Convert text to DataFrame
    csv_data = StringIO(response.text)
    # df = pd.read_csv(csv_data)
    df = pd.read_csv(csv_data)
    print(df.head())
else:
    print(f"Failed to fetch data. Status code: {response.status_code}")
```

```python
# simulation d'un flux de données en temps réel
while True:
    dict_stock = df.sample(1).to_dict(orient="records")[0]
    producer.send('demo_01', value=dict_stock)
```
